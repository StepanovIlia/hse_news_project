```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

url_ = 'https://www.vedomosti.ru/archive/'
res = pd.DataFrame()
t0 = time.time()

    
for y in range(2014,2020):
    
        for m in range(1,13):
            
            for d in range(1,32):

                    st_m = str(m)
                    if len(st_m) == 1: st_m = '0'+st_m

                    st_d = str(d)
                    if len(st_d) == 1: st_d = '0'+st_d

                    st = str(y) + '/' + st_m + '/' + st_d

                    url = url_ + st

                    try: 
                        page = requests.get(url)
                    except:
                        r = {'step':st, 'error':'downloading error', 'time':(time.time()- t0)}
                        res = res.append(pd.DataFrame(r, index = [len(res)]))
                        continue

                    try:
                        p = BeautifulSoup(page.content, "html.parser")
                        url_list = p.find_all('li', {'class':'b-article b-article_list'})
                    except:
                        r = {'step':st, 'error':'parsing error', 'time':(time.time()- t0)}
                        res = res.append(pd.DataFrame(r, index = [len(res)]))


                    for x in url_list:
                        r = {'step':st, 'url':x.find('a').get('href'), 'time':(time.time()- t0)}
                        res = res.append(pd.DataFrame(r, index = [len(res)]))

                    print(st, end = '\r')



    #res.to_excel('res_40_2018.xlsx')
```

```python
from tqdm import tqdm
```

```python
res = pd.read_excel('texts/vedomosti_r.xlsx')
```

```python
url_list = res.copy()
```

```python


url_ = 'https://www.vedomosti.ru'
#res = pd.read_excel('texts.xlsx')
res = pd.DataFrame()

urls = url_list['url'].tolist()



for x_ in tqdm(range(0, 29811)):
    
    x = urls[x_]
    url = url_ + x

    try:
        page = requests.get(url).content
    except:
        r = {'url': x, "error": "downloading error"}
        res = res.append(pd.DataFrame(r, index=[len(res)]))
        continue

    p = BeautifulSoup(page, "html.parser")
    # текст статьи
    try:
        p_ = p.find_all('div', {'class':'b-document__left'})[0]
        st = ''
        for x in p_.find_all('p'): st = st + x.text + ' '
    except:
        st = None
    
    
    try: 
        article_header = p.find('h1', {'class':"title"}).text.replace('\n', '')
    except:
        article_header = None
    
    try:
        article_subheader = p.find('div', {'class':'subtitle'}).text.replace('\n', '')
    except:
        article_subheader = None


    r = {'url': url, "text":st, "article_header":article_header, "artical_subheader":article_subheader}
    res = res.append(pd.DataFrame(r, index = [len(res)]))
    
    
```

```python
res.to_excel('texts/vedomosti_1.xlsx', index = False)
```
